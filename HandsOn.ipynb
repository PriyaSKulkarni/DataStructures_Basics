```python
import numpy as np
import matplotlib.pyplot as plt
import timeit

def f(n):
    x = 1
    for i in range(1, n+1):
        for j in range(1, n+1):
            x = x + 1
    return x

# Timing the function for various values of n using timeit
n_values = np.arange(1, 1001, 10)
times = np.zeros(len(n_values))

for i, n in enumerate(n_values):
    time_taken = timeit.timeit(lambda: f(n), number=10)  # Adjust the number of repetitions as needed
    times[i] = time_taken / 10  # Average time per function call

# Plotting time vs n
plt.figure()
plt.plot(n_values, times, 'o-', linewidth=2, label='Actual Data')
plt.xlabel('n')
plt.ylabel('Time (s)')
plt.title('Algorithm Runtime Analysis')
plt.grid(True)

# Fitting a curve (polynomial) to the data
p = np.polyfit(n_values, times, 2)
fitted_curve = np.polyval(p, n_values)

plt.plot(n_values, fitted_curve, '--', linewidth=2, label='Fitted Curve')

# Finding upper and lower bounds (polynomials)
upper_bound = 1.1 * np.polyval(p, n_values)  # Adjust the constant factor as needed
lower_bound = 0.9 * np.polyval(p, n_values)  # Adjust the constant factor as needed

# Plotting upper and lower bounds
plt.plot(n_values, upper_bound, ':', linewidth=1.5, label='Upper Bound')
plt.plot(n_values, lower_bound, ':', linewidth=1.5, label='Lower Bound')

# Finding n_0 (zoom in on the plot)
n_0 = 100  # Adjust this value based on the plot

# Marking the approximate location of n_0
plt.scatter(n_0, np.polyval(p, n_0), s=50, c='red', marker='s', label=r'$n_0$')
plt.text(n_0, np.polyval(p, n_0), r' $n_0$', ha='right', va='bottom')

plt.legend()
plt.show()
```
